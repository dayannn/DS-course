```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

onehot_encoder = OneHotEncoder(sparse=False)

encoded_categorical_columns = pd.DataFrame(onehot_encoder.fit_transform(dataset[['Evaporation']]))
encoded_categorical_columns.head()

# ct = ColumnTransformer(
#     [('oh_enc', OneHotEncoder(sparse=False), ['Evaporation', 'Sunshine', 'WindDir9am', 'RainToday']),],  
#     remainder='passthrough')
# print(ct.fit_transform(X)) # Notice the output is a string
```

```python
Sydney.loc[Sydney['RainTomorrow'] == 'Yes', 'RainTomorrow'] = 1
Sydney.loc[Sydney['RainTomorrow'] == 'No', 'RainTomorrow'] = 0
Sydney[:10]
```

```python
X, y = Sydney.iloc[:, :-1], Sydney.iloc[:, -1:]
```

```python
y[y=='No']
```

```python

X.drop(inplace=True, columns=['Date', 'Location'])
X.drop(columns=['Sunshine', 'Evaporation'], inplace=True)
X = pd.get_dummies(X)
X.dropna(inplace=True)
X
```

```python
model = KMeans(n_clusters=2)

model.fit(X)
predictions = model.predict(X)

predictions = 1 - predictions

# predictions[predictions == 1] = 50
# predictions[predictions == 0] = 1
# predictions[predictions == 3] = 0
# predictions[predictions == 50] = 0

print(predictions)
```

```python
true_pred = y.loc[X.index.tolist()]
```

```python
true_pred[true_pred == 'No'] = 0
true_pred[true_pred == 'Yes'] = 1
```

```python
true_pred = true_pred.values.reshape([-1])
```

```python
true_pred = np.array(true_pred, dtype=int)
```

```python
from sklearn import metrics
from sklearn.metrics.cluster import adjusted_rand_score

print ("accuracy = ", metrics.accuracy_score(y_true=true_pred, y_pred=predictions))
print ("precision = ", metrics.precision_score(y_true=true_pred, y_pred=predictions))
print ("recall = ", metrics.recall_score(y_true=true_pred, y_pred=predictions))
print ("f1 = ", metrics.f1_score(y_true=true_pred, y_pred=predictions))
```
